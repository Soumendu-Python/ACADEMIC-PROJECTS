```python
import pandas as pd
from bs4 import BeautifulSoup
from requests import get
```

```python
url='https://www.snapdeal.com/search?keyword=headphone&santizedKeyword=&catId=&categoryId=0&suggested=false&vertical=&noOfResults=20&searchState=&clickSrc=go_header&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=&url=&utmContent=&dealDetail=&sort=rlvncy'
page=get(url)
print(page.text)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
res=soup.find('div',class_='comp comp-right-wrapper ref-freeze-reference-point clear')
res.text
```

```python
result=res.find('p',class_='product-title')
result.text
```

```python
res1=res.find('span',class_='lfloat product-price')
res1.text
```

```python
res2=res.find('div',class_='product-discount')
res2.text
```

```python
res5=soup.find_all('div',class_='product-discount')
for r in res5:
    print(r.text.strip())
```

```python
res3=res.find('p',class_='product-rating-count')
res3.text
```

```python
res4=res.find('span',class_='lfloat product-desc-price strike')
res4.text
```

```python
hname=[]
price=[]
op=[]
rate=[]
dis=[]
```

```python
hn=soup.find_all('p',class_='product-title')
for h in hn:
    print(h.text)
    hname.append(h.text)
```

```python
hname
```

```python
pr=soup.find_all('span',class_='lfloat product-price')
for p in pr:
    print(p.text)
    price.append(p.text)
```

```python
orpr=soup.find_all('span',class_='lfloat product-desc-price strike')
for o in orpr:
    op.append(o.text)
```

```python
r=soup.find_all('p',class_='product-rating-count')
for ra in r:
    rate.append(ra.text)
```

```python
d=soup.find_all('div',class_='product-discount')
for di in d:
    print(di.text.strip())
    dis.append(di.text.strip())
```

```python
print(len(hname))
print(len(price))
print(len(orpr))
print(len(dis))
print(len(rate))
```

```python
rate.append(0)
```

```python
headphones=pd.DataFrame({'Model':hname,'Ratings':rate,'Original Price':op,'Discount Percentage':dis,
                        'Exclusive Price':price})
headphones
```
